```python
# converting the reservation_status_date to date
# as we can see in above information the reservation_status_data is in object format

df['reservation_status_date'] = pd.to_datetime(df.reservation_status_date, infer_datetime_format=True)
```

```python
for col in df.select_dtypes(include='O').columns:
    print(col+':')
    print(df[col].unique())
    print('-'*50)
```

```python
df.isna().sum()
```

```python
# We have to drop company and agent columns because 
# We don't need agent data and company's missing values are too large to handle.

df.drop(['agent', 'company'], axis=1, inplace=True)

# We have to remove the rows having missing value in country and children because
# missing data is very less compare to the size of data, it will not effect the analysis of the data.

df.dropna(inplace=True)
```

### :: Getting the summary of the numerial columns

```python
df.describe()
```

```python
sns.boxplot(data=df.adr, orient='h')
```

```python
sns.scatterplot(data=df.adr)
```

```python
# As we can see there are outliers in the adr field.
# This is the required field for our analysis so we are going to find and remove the outliers.
# As we can see all the data is in cluster except some dots above 5000.
# So we can remove those values.

df = df[df.adr < 5000]
```

```python
# As we can see now the dat is cleaner now.
sns.scatterplot(data=df.adr)
```

### :: Data Analysis and Visualisation

```python
# Lets find out the cancellation percentage.

df.is_canceled.value_counts(normalize=True)

plt = sns.barplot(data=df, y=df.is_canceled.value_counts(), x=['Not Canceled','Canceled'])
plt.set(ylabel='value count')
```

```python
# let's visualize the number of cancelation based on hotels

plt.figure(figsize=(8,6))
ax = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='pastel')
ax.set(xlabel='hotels', ylabel='Number of reservations', title="Observation based on Hotels")
plt.legend(['Not Canceled', 'canceled'])
plt.show()
```

```python
# Check cancellation percent vased on hotels

print('Resort Hotel')
res_hotel = df[df.hotel == 'Resort Hotel']
res_hotel.is_canceled.value_counts(normalize='True')
```

```python
print('City Hotel')
city_hotel = df[df.hotel == 'City Hotel']
city_hotel.is_canceled.value_counts(normalize='True')
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=res_hotel.groupby('reservation_status_date').adr.mean())
sns.lineplot(data=city_hotel.groupby('reservation_status_date').adr.mean())
plt.legend(['Resort Hotel', 'City Hotel'])
plt.ylabel('Average Daily Rate/Price')
plt.show()
```

```python
# Let's find the reservation and cancelation based on months

df['month'] = df.reservation_status_date.dt.month
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='month', hue='is_canceled', data=df)
plt.legend(['Not Canceled', 'Canceled'])
plt.ylabel('Number of reservations')
plt.show()
```

```python
# Let's find out the adr based on the month

plt.figure(figsize=(12, 6))
sns.barplot(x='month', y='adr', data=df[df.is_canceled == 1].groupby('month')[['adr']].sum().reset_index())
plt.ylabel('Average Daily Rate/Price')
plt.show()
```

```python
# By checking above two charts we can find out that the higher is the price, higher is the cancelation rate
# Like in 1st month the ard is more than 500000 and in the cancelation rate is hightest in the 1st month as well.
```

```python
# lets find out top 10 countries having most cancelation rate.

top_ten_country = data=df[df.is_canceled == 1].country.value_counts()[:10]
plt.figure(figsize=(10,7))
plt.pie(top_ten_country, autopct='%.2f', labels=top_ten_country.index, explode=[0.03,0.09,0.08,0.07,0.06,0.05,0.04,0.03,0.02,0.01])
plt.show()
```

```python
# Portugal are having most cancelation rate of 70%

# lets find out from which mean the clients are reserving the hotels

df.market_segment.value_counts()
```

```python
# we can also show the percent of the above observation

df.market_segment.value_counts(normalize=True)
```

```python
# Lets find out the cancelation rate for above observations.

df[df.is_canceled == 1].market_segment.value_counts(normalize=True)
```

```python
# We can see the cancelation rate is also higher for online travel agents.
```

```python
cancelation_df_adr = df[df.is_canceled == 1].groupby('reservation_status_date')[['adr']].mean()
cancelation_df_adr.reset_index(inplace=True)
cancelation_df_adr.sort_values('reservation_status_date', inplace=True)


non_cancelation_df_adr = df[df.is_canceled == 0].groupby('reservation_status_date')[['adr']].mean()
non_cancelation_df_adr.reset_index(inplace=True)
non_cancelation_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(18,10))
sns.lineplot(data=cancelation_df_adr, palette='magma', label='Canceled')
sns.lineplot(data=non_cancelation_df_adr, palette='twilight', label='Non Canceled')
plt.show()
```
